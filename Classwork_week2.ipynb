```python
# Generate 100 random integer ages between 1 and 99
ages = np.random.randint(1, 100, 100)
print (ages)
plt.hist(ages, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of ages')
plt.show()
```

```python
#binning age in age groups
bins = [0, 18, 35, 50, float('inf')]
labels = ['0-18', '19-35', '36-50', '50+']
df_ages = pd.DataFrame({'Age': ages})
df_ages['Age_Group'] = pd.cut(df_ages['Age'], bins=bins, labels=labels, right=True, include_lowest=True) #this was unique
print(df_ages.head())
#distribution of age dataset from data frame
plt.hist (df_ages['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()

#distribution of ages in groups
sns.countplot(x='Age_Group', data=df_ages)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Age Groups')
plt.show()
```

```python
np.random.seed(42)

samples= np.random.normal(0,1,100)
print (samples)
plt.hist(samples, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Samples')
plt.show()
```

```python
# Plot a boxplot on the y-axis
sns.boxplot(y=samples)
plt.title('Boxplot of Samples ')
plt.show()
```

```python
plt.scatter(samples, samples)
plt.xlabel('Sample Value')
plt.ylabel('Sample Value')
plt.title('Scatter Plot of Samples')
plt.show()
```

```python
# Define the categories
categories = ['Category A', 'Category B', 'Category C']

# Randomly assign each sample to a category
# We'll use np.random.choice to pick a category for each sample
categorical_samples = np.random.choice(categories, size=len(samples))

# Create a DataFrame to see the original samples and their assigned categories
df_samples_categorized = pd.DataFrame({
    'Sample Value': samples,
    'Category': categorical_samples
})

# Display the first few rows
print(df_samples_categorized.head())
```

```python
pieplot = df_samples_categorized['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
pieplot.set_title('Distribution of Categories')
plt.show()
```
